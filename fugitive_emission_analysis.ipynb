## FUGITIVE EMISSION DATA ANALYSIS


```python
#installating and importing libraries, dependecies

#! pip install pandas
#! pip install openpyxl # this is a library to read xlsx file

import pandas as pd
import os

# this is to Suppress the warning
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")
```

```python
#importing the excel data to jupiter notebook

file_paths = [
    r"./Fugitive emissions_MY_KL.xlsx",
    r"./Fugitive emissions_PH_CEB.xlsx",
    r"./Fugitive emissions_PH_MNL.xlsx",
    r"./Fugitive emissions_SG_HQ.xlsx",
    r"./Fugitive emissions_SG_Tuas.xlsx",
    # Add more file paths as needed
]

# Initializing an empty list to store DataFrames
data_frames = []

# Looping through the file paths and read each Excel file into a DataFrame
for file in file_paths:
    df = pd.read_excel(file)
    data_frames.append(df)

# Combining all the DataFrames into one
combined_df = pd.concat(data_frames, ignore_index=True)

print("Files combined successfully!")
```

```python
# cheking the information of the combined data
combined_df.info()
```

```python
# checking the number of rows and columns of the dataframe
combined_df.shape
```

```python
# displaying the dataframe
combined_df
```

```python
# dropping the columns that are not nedeed in the analysis
new_df = combined_df.drop(columns=["Description",
                                   "Consumption End",
                                   "Data Quality Type",
                                   "Emission Factor Library",
                                   "Is Market Based",
                                   "Industrial Process Type",
                                   "Leak Source",
                                   "Leak Description",
                                   "CO2 Unit",
                                   "CH4 Unit",
                                   "SF6 Unit",
                                   "NF3 Unit",
                                   "N2O Unit",
                                  "Other GHGs Unit",
                                  "PFCs Unit",
                                  "HFCs Unit",
                                  "Emission Rate",
                                  "CO2 Percentage",
                                  "CH4 Percentage",
                                  "N2O Percentage",
                                  "HFCs Percentage",
                                  "PFCs Percentage",
                                  "SF6 Percentage",
                                  "NF3 Percentage",
                                  "Other GHGs Percentage",
                                   ])

new_df
```

```python
# checking for data types and null values before doing aggregation 
new_df.info()
```

```python
# renaming the columns 
final_df = new_df.rename(columns={
            "Name": "Leak Asset ID", 
            "Consumption Start": "Consumption Period"
})

# adding column for the total GHGs
final_df['Total GHGs'] = final_df[['CO2', 'CH4', 'N2O', 'HFCs', 'PFCs', 'SF6','NF3', 'Other GHGs']].sum(axis=1)

# displaying the final dataframe the will be used for analysis
final_df
```

```python
#exporting the final dataframe to excel, ready for visualisation 
final_df.to_excel('fugitive_emission_final_data.xlsx', sheet_name='data', index=False)
```
